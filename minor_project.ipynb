Title:- Data Visualization and Styling in Python

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [120, 180, 150, 210, 240]

plt.figure(figsize=(8,5))

# Line style customization
plt.plot(months, sales, color='green', linestyle='--', marker='o', linewidth=2, label='Sales')

# Titles and labels
plt.title("Monthly Sales Trend", fontsize=16, color='darkblue', pad=20)  # 'pad' adds spacing
plt.xlabel("Month", fontsize=12)
plt.ylabel("Revenue ($)", fontsize=12)
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)

# Annotation - move slightly higher to prevent overlap
plt.annotate(
    'Highest Sales',
    xy=('May', 240),          # actual data point
    xytext=('Apr', 270),      # moved text higher and to the left
    arrowprops=dict(facecolor='black', arrowstyle='->'),
    fontsize=10,
    color='black'
)

plt.show()
```

```python
categories = ['Store A', 'Store B', 'Store C']
revenue = [350, 500, 420]

sns.set_theme(style="darkgrid")
plt.figure(figsize=(7,5))
plt.bar(categories, revenue, color=['#5DADE2', '#58D68D', '#F5B041'])
plt.title("Store Revenue Comparison", fontsize=14)
plt.xlabel("Stores")
plt.ylabel("Revenue ($)")
plt.grid(axis='y')

# Annotation
for i, val in enumerate(revenue):
    plt.text(i, val + 10, f"${val}", ha='center', fontsize=10)

plt.show()
```

```python
import numpy as np
sns.set_theme(style="whitegrid")

data = np.random.normal(70, 10, 200)

plt.figure(figsize=(8,5))
plt.hist(data, bins=10, color='skyblue', edgecolor='black')
plt.title("Distribution of Exam Scores", fontsize=14)
plt.xlabel("Score")
plt.ylabel("Frequency")

# Annotation
plt.axvline(np.mean(data), color='red', linestyle='dashed', linewidth=2)
plt.text(np.mean(data)+1, 20, 'Mean', color='red')

plt.show()
```

```python
np.random.seed(10)
ad_spend = np.random.randint(10, 100, 30)
sales = ad_spend * 2 + np.random.randint(0, 20, 30)

sns.set_theme(style="ticks")
plt.figure(figsize=(8,5))
plt.scatter(ad_spend, sales, color='teal', marker='o', edgecolor='black')
plt.title("Ad Spend vs Sales", fontsize=14)
plt.xlabel("Advertising Spend ($)")
plt.ylabel("Sales ($)")
plt.grid(True)

# Annotation
plt.annotate('Trend Line', xy=(60,130), xytext=(40,160),
             arrowprops=dict(facecolor='black', arrowstyle='->'))

plt.show()
```

```python
sizes = [40, 30, 20, 10]
labels = ['Marketing', 'R&D', 'Operations', 'HR']
colors = ['#5DADE2', '#58D68D', '#F5B041', '#EC7063']

plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors,
        startangle=90, explode=[0.05, 0, 0, 0])
plt.title("Company Budget Allocation", fontsize=14)
plt.show()
```

```python
data = sns.load_dataset("tips")

sns.set_theme(style="whitegrid")

plt.figure(figsize=(8,5))
sns.boxplot(x="day", y="total_bill", hue="day", data=data, palette="pastel", legend=False)
plt.title("Total Bill by Day", fontsize=14)
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill ($)")
plt.show()
```

```python
corr = data.corr(numeric_only=True)
plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True, cmap='YlGnBu')
plt.title("Correlation Heatmap", fontsize=14)
plt.show()
```

```python
sns.set_theme(style="ticks")
sns.pairplot(data, hue="sex", palette="husl")
plt.suptitle("Pairwise Relationships in Tips Dataset", y=1.02, fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.set_theme(style="whitegrid")
sns.violinplot(x="day", y="total_bill", hue="day", data=data, palette="pastel", legend=False)
plt.title("Violin Plot of Total Bill by Day", fontsize=14)
plt.xlabel("Day")
plt.ylabel("Total Bill ($)")
plt.show()
```
